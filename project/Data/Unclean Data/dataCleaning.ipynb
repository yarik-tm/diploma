```python
import pandas as pd
import numpy as np
```

```python
# Import all data files

df_av_temp = pd.read_json('country-yearly-average-temperature.json')

df_2015 = pd.read_csv('2015.csv')

df_2016 = pd.read_csv('2016.csv')

df_2017 = pd.read_csv('2017.csv')

df_2018 = pd.read_csv('2018.csv')

df_2019 = pd.read_csv('2019.csv')

df_2020 = pd.read_csv('2020.csv')

df_2021 = pd.read_csv('2021.csv')
```

```python
# Add year column to each dataframe

df_2015['Year'] = '2015'

df_2016['Year'] = '2016'

df_2017['Year'] = '2017'

df_2018['Year'] = '2018'

df_2019['Year'] = '2019'

df_2020['Year'] = '2020'

df_2021['Year'] = '2021'
```

```python
# Review column titles for each df and ensure consistency. Use df_2015 as standard.

# dropping standard error as this column likely won't be displayed in visuals
df_2015c = df_2015.drop(['Standard Error', 'Dystopia Residual', 'Family','Region'], axis=1).rename(columns ={
                                                                    'Country': 'country',
                                                                    'Happiness Rank': 'happiness_rank',
                                                                    'Happiness Score': 'happiness_score',
                                                                    'Economy (GDP per Capita)': 'economy_gdp_per_capita',
                                                                    'Happiness Score': 'happiness_score',
                                                                    'Health (Life Expectancy)': 'health_life_expectancy',
                                                                    'Freedom': 'freedom',
                                                                    'Trust (Government Corruption)': 'trust_government_corruption',
                                                                    'Generosity': 'generosity',
                                                                    'Year': 'year'})

df_2016c = df_2016.drop(['Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual',], axis = 1).rename(columns ={
                                                                    'Country': 'country',
                                                                    'Happiness Rank': 'happiness_rank',
                                                                    'Happiness Score': 'happiness_score',
                                                                    'Economy (GDP per Capita)': 'economy_gdp_per_capita',
                                                                    'Happiness Score': 'happiness_score',
                                                                    'Health (Life Expectancy)': 'health_life_expectancy',
                                                                    'Freedom': 'freedom',
                                                                    'Trust (Government Corruption)': 'trust_government_corruption',
                                                                    'Generosity': 'generosity',
                                                                    'Year': 'year'})

df_2017c= df_2017.drop(['Whisker.high',
                        'Whisker.low',
                        'Dystopia.Residual',
                        'Family'], axis=1).rename(columns={'Happiness.Rank': 'happiness_rank',
                                                                        'Happiness.Score': 'happiness_score',
                                                                        'Economy..GDP.per.Capita.': 'economy_gdp_per_capita',
                                                                        'Health..Life.Expectancy.': 'health_life_expectancy',
                                                                        'Trust..Government.Corruption.': 'trust_government_corruption'})
df_2018c = df_2018.drop(['Social support'], axis=1).rename(columns={'Overall rank': 'happiness_rank',
                                                                   'Country or region': 'country',
                                                                   'Score': 'happiness_score',
                                                                   'GDP per capita': 'economy_gdp_per_capita',
                                                                   'Healthy life expectancy': 'health_life_expectancy',
                                                                   'Freedom to make life choices': 'freedom',
                                                                   'Perceptions of corruption': 'trust_government_corruption'})

df_2019c = df_2019.drop(['Social support'], axis=1).rename(columns={'Overall rank': 'happiness_rank',
                                                                   'Country or region': 'country',
                                                                   'Score': 'happiness_score',
                                                                   'GDP per capita': 'economy_gdp_per_capita',
                                                                   'Healthy life expectancy': 'health_life_expectancy',
                                                                   'Freedom to make life choices': 'freedom',
                                                                   'Perceptions of corruption': 'trust_government_corruption'})

# Need inverse log of 'Logged GDP per capita'
# Need to add Happiness rank
df_2020c = df_2020.drop(['upperwhisker','lowerwhisker',
                         'Ladder score in Dystopia',
                         'Standard error of ladder score',
                         'Explained by: Log GDP per capita',
                        'Explained by: Social support',
                        'Explained by: Healthy life expectancy',
                        'Explained by: Freedom to make life choices',
                        'Explained by: Generosity',
                        'Explained by: Perceptions of corruption',
                        'Dystopia + residual',
                        'Social support',
                        'Regional indicator'], axis=1).rename(columns ={
                                                                    'Country name': 'country',
                                                                    'Logged GDP per capita': 'economy_gdp_per_capita',
                                                                    'Ladder score': 'happiness_score',
                                                                    'Healthy life expectancy': 'health_life_expectancy',
                                                                    'Freedom to make life choices': 'freedom',
                                                                    'Perceptions of corruption': 'trust_government_corruption'})

df_2021c = df_2021.drop(['upperwhisker','lowerwhisker',
                         'Ladder score in Dystopia',
                         'Standard error of ladder score',
                         'Explained by: Log GDP per capita',
                        'Explained by: Social support',
                        'Explained by: Healthy life expectancy',
                        'Explained by: Freedom to make life choices',
                        'Explained by: Generosity',
                        'Explained by: Perceptions of corruption',
                        'Dystopia + residual',
                        'Social support',
                        'Regional indicator'], axis=1).rename(columns ={
                                                                    'Country name': 'country',
                                                                    'Logged GDP per capita': 'economy_gdp_per_capita',
                                                                    'Ladder score': 'happiness_score',
                                                                    'Healthy life expectancy': 'health_life_expectancy',
                                                                    'Freedom to make life choices': 'freedom',
                                                                    'Perceptions of corruption': 'trust_government_corruption'})
```

```python
# Make all columns names lowercase

df_2015c.columns = df_2015c.columns.str.lower()

df_2016c.columns = df_2016c.columns.str.lower()

df_2017c.columns = df_2017c.columns.str.lower()

df_2018c.columns = df_2018c.columns.str.lower()

df_2019c.columns = df_2019c.columns.str.lower()

df_2020c.columns = df_2020c.columns.str.lower()

df_2021c.columns = df_2021c.columns.str.lower()
```

```python
# add Happiness Rank to 2020, 2021 data
# Reorganize columns


df_2016c = df_2016c[['country',
 'happiness_rank',
 'happiness_score',
 'economy_gdp_per_capita',
 'health_life_expectancy',
 'freedom',
 'trust_government_corruption',
 'generosity',
 'year']]

df_2017c = df_2017c[['country',
 'happiness_rank',
 'happiness_score',
 'economy_gdp_per_capita',
 'health_life_expectancy',
 'freedom',
 'trust_government_corruption',
 'generosity',
 'year']]

df_2018c = df_2018c[['country',
 'happiness_rank',
 'happiness_score',
 'economy_gdp_per_capita',
 'health_life_expectancy',
 'freedom',
 'trust_government_corruption',
 'generosity',
 'year']]

df_2019c = df_2019c[['country',
 'happiness_rank',
 'happiness_score',
 'economy_gdp_per_capita',
 'health_life_expectancy',
 'freedom',
 'trust_government_corruption',
 'generosity',
 'year']]


df_2020c.sort_values(by='happiness_score', ascending=True)
df_2020c['happiness_rank'] = range(1, len(df_2020c)+1)

df_2020c = df_2020c[['country',
 'happiness_rank',
 'happiness_score',
 'economy_gdp_per_capita',
 'health_life_expectancy',
 'freedom',
 'trust_government_corruption',
 'generosity',
 'year']]

df_2021c.sort_values(by='happiness_score', ascending=True)
df_2021c['happiness_rank'] = range(1, len(df_2021c)+1)

df_2021c = df_2021c[['country',
 'happiness_rank',
 'happiness_score',
 'economy_gdp_per_capita',
 'health_life_expectancy',
 'freedom',
 'trust_government_corruption',
 'generosity',
 'year']]
```

```python
# 2021 & 2020 Health (Life Expectancy) needs to be normalized

df_2021c['health_life_expectancy'] = (df_2021c['health_life_expectancy'] - df_2021c['health_life_expectancy'].min()) / (df_2021c['health_life_expectancy'].max() - df_2021c['health_life_expectancy'].min())

df_2020c['health_life_expectancy'] = (df_2020c['health_life_expectancy'] - df_2020c['health_life_expectancy'].min()) / (df_2020c['health_life_expectancy'].max() - df_2020c['health_life_expectancy'].min())

# 2021 & 2020 Economy (GDP per Capita)

# df_2021c['Economy (GDP per Capita)'] = np.exp(df_2021c['Economy (GDP per Capita)']) / 10
```

```python
# Concat all df
clean_df = pd.concat([df_2015c, df_2016c, df_2017c, df_2018c, df_2019c, df_2020c, df_2021c]).reset_index(drop=True)
```

```python
# Check for NA values

if len(df_av_temp) - len(df_av_temp.dropna()) == 0:
    print('No Na values in temp data')
    
if len(clean_df) - len(clean_df.dropna()) == 0:
    print('No NA value in happiness data')
else:
    clean_df.dropna(inplace=True)
    
# Remove units from temp value

split = df_av_temp['yearly-average-temp'].str.split(' ', expand = True)

df_av_temp['av_temp'] = split[0]
    
df_av_temp.drop(columns=['yearly-average-temp'], inplace=True)

# df_av_temp.head()
```

```python
# Confirm both dataframes have the same countries. This will be required for SQL.

for index,row in clean_df.iterrows():
    try:
        df_av_temp.loc[df_av_temp['country'] == row['country']]
#         print(df_av_temp.loc[df_av_temp['country'] == row['Country']])
    except:
        print(f'{row} not in clean data df')
```

```python
# Set indexes to countries

df_av_temp.set_index(['country'], inplace=True)

clean_df.set_index(['country'], inplace=True)
```

```python
# Send cleaned dataframes to CSV

clean_df.to_csv('../Clean Data/happinessData.csv')

df_av_temp.to_csv('../Clean Data/avTempData.csv')
```
